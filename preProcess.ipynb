# 1. 폴더안의 이미지 해상도와 포맷 변경

```python
import os
import cv2

def resize_and_convert_format(root_folder, target_ratio=0.1, target_format='.png'):
    for foldername, subfolders, filenames in os.walk(root_folder):
        for filename in filenames:
            file_path = os.path.join(foldername, filename)

            # 영상 파일인지 확인
            if file_path.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                try:
                    # 영상 파일 읽기
                    img = cv2.imread(file_path)

                    # 원본 이미지의 가로 세로 크기 가져오기
                    height, width, _ = img.shape

                    # 새로운 해상도 계산 (원본 해상도의 비율을 유지)
                    new_width = int(width * target_ratio)
                    new_height = int(height * target_ratio)

                    # 영상 파일의 해상도 변경
                    img_resized = cv2.resize(img, (new_width, new_height))

                    # 새로운 파일 경로 생성
                    new_folder_path = os.path.join(foldername, 'converted')
                    os.makedirs(new_folder_path, exist_ok=True)  # "converted" 디렉터리 생성
                    new_file_path = os.path.join(new_folder_path, f"{os.path.splitext(filename)[0]}.png")  # .png로 고정

                    # 영상 파일의 포맷 변경 및 저장
                    cv2.imwrite(new_file_path, img_resized)

                    print(f"Processed: {file_path} -> {new_file_path}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")


```

```python
# 시작 폴더 지정 (본인의 경우에 맞게 수정)
# root_folder = '/Users/giy/Desktop/09_copy/'
root_folder = DATA_PATH

# 함수 호출 (기본값은 원본 해상도의 50%로 축소, 포맷은 JPEG)
resize_and_convert_format(root_folder)
```

```python

# frames_to_cropped_movie(./data/kor/01/훌륭하게습득했다)
from lipnet.lipreading.videos import Video
import os, fnmatch, sys, errno  
from skimage import io
import numpy as np

FACE_PREDICTOR_PATH = './predictors/shape_predictor_68_face_landmarks.dat'

def frames_to_cropped_movie(folder_path: str):
    for foldername, subfolders, filenames in os.walk(root_folder):

        if 'converted' in foldername:
            
            print(foldername)
            video = Video(vtype='face', face_predictor_path=FACE_PREDICTOR_PATH).from_frames(foldername)

            target_dir = os.path.join(foldername, '..', 'cropped')
            os.makedirs(target_dir, exist_ok=True)  # "cropped" 디렉터리 생성

            i = 0
            for frame in video.mouth:
                frame = (frame * 255).astype(np.uint8)
                # print(frame)
                io.imsave(os.path.join(target_dir, "mouth_{0:03d}.png".format(i)), frame)
                i += 1
            print ("Saved at: {}".format(target_dir))
```

```python
# 파일의 개수에 따라 복사해 그 수를 늘리기
import os
import shutil

def double_files_count(root_folder, target_count=75):
    for foldername, subfolders, filenames in os.walk(root_folder):

        if 'cropped' in foldername and 'cropped_video' not in foldername:
            current_file_count = len(filenames)
            
            if current_file_count < target_count // 3:
                num_copies = 2
            elif current_file_count < target_count // 2:
                num_copies = 1
            else:
                num_copies = 0  # 필요한 복사 횟수를 초과하는 경우

            for filename in filenames:
                source_file_path = os.path.join(foldername, filename)
                print(source_file_path)
                
                for i in range(num_copies):  # 2번 반복
                    # 파일을 복사하여 새로운 파일 생성
                    destination_file_path = os.path.join(foldername, f"{filename.split('.')[0]}_copy_{i}.{filename.split('.')[-1]}")
                    shutil.copy2(source_file_path, destination_file_path)

                    print(f"Copied: {destination_file_path}")


```

# 4. 파일 수 75로 맞추기

```python
import os
import cv2
import shutil

def resize_image_count(root_folder, target_count=75, target_ratio=0.1):
    for foldername, subfolders, filenames in os.walk(root_folder):

        if 'cropped' in foldername:
            
            # 폴더 내 이미지 파일 수 확인
            image_count = len([filename for filename in filenames if filename.lower().endswith(('.png'))])
            
            if image_count != 0 and image_count < target_count:
                # 부족한 이미지 수 계산
                images_to_add = target_count - image_count
                
                # 이미지 파일 이름 순으로 정렬
                sorted_filenames = sorted(filenames)
                
                # 이미지 파일 추가
                for i in range(images_to_add):
                    source_file_path = os.path.join(foldername, sorted_filenames[-1])  # 맨 마지막의 이미지 반복
                    destination_folder_path = os.path.join(foldername, '..', 'cropped_padded')
                    os.makedirs(destination_folder_path, exist_ok=True) 
                    destination_file_path = os.path.join(destination_folder_path, f"mouth_{i + 1:03d}.png")
                    
                    shutil.copy2(source_file_path, destination_file_path)
                    print(f"Added: {destination_file_path}")

                # 기존 이미지도 이름을 변경하여 복사
                for i in range(image_count):
                    source_file_path = os.path.join(foldername, sorted_filenames[i])  
                    destination_file_path = os.path.join(foldername, '..', 'cropped_padded', f"mouth_{images_to_add + i + 1:03d}.png")

                    shutil.copy2(source_file_path, destination_file_path)
                    print(f"Copied: {destination_file_path}")

            elif image_count > target_count:
                # 초과한 이미지 수 계산
                images_to_remove = image_count - target_count
                
                # 이미지 파일 이름 순으로 정렬
                sorted_filenames = sorted(filenames)
                
                # 초과한 이미지 파일 제거
                for i in range(images_to_remove):
                    destination_folder_path = os.path.join(foldername, '..', 'cropped_padded')
                    os.makedirs(destination_folder_path, exist_ok=True) 
                    file_to_remove = os.path.join(destination_folder_path, sorted_filenames[-(i + 1)])
                    
                    os.remove(file_to_remove)
                    print(f"Removed: {file_to_remove}")


```

```python
# 시작 폴더 지정 (본인의 경우에 맞게 수정)
# root_folder = '/Users/giy/Desktop/09_copy/'
root_folder = DATA_PATH 

# 함수 호출 (기본값은 원본 해상도의 10%로 축소, 이미지 수를 75장으로 맞춤)
resize_image_count(root_folder)
```

# 5. 영상으로 만든다

```python
import cv2
import os
from pathlib import Path


FPS = 25
def makeFramesToVideo(root_folder):
    for foldername, subfolders, filenames in os.walk(root_folder):
        
        if 'cropped_padded' in foldername:

            # subfolder 내의 모든 .png 파일 가져오기 및 정렬
            image_files = sorted([filename for filename in filenames if filename.lower().endswith(".png")])
            # print(image_files)

            if len(image_files) == 0:
                continue  # .png 파일이 없는 경우 건너뛰기

            # 첫 번째 이미지를 기준으로 영상 크기 가져오기
            first_image = cv2.imread(os.path.join(foldername, image_files[0]))
            height, width, _ = first_image.shape

            # 비디오 파일 생성
            parent_video_path = os.path.join(foldername, '..', 'cropped_video')
            # parent_video_path = os.path.abspath(os.path.join(folder_path, '..', 'cropped_video'))
            os.makedirs(parent_video_path, exist_ok=True) 
            video_file_name = foldername.split('/')[-2]
            video_file_name = "".join(video_file_name.split())
            print(video_file_name)

            video_path = os.path.join(parent_video_path, f"{video_file_name}.mp4")
            video_writer = cv2.VideoWriter(video_path, cv2.VideoWriter_fourcc(*'MP4V'), FPS, (width, height))

            # 각 이미지를 비디오에 추가
            for image_file in image_files:
                image_path = os.path.join(foldername, image_file)
                # print(image_path)
                frame = cv2.imread(image_path)
                video_writer.write(frame)

            # 비디오 파일 닫기
            video_writer.release()
            print(f'{video_path} 파일 생성이 완료되었습니다.')




```

```python
import os
import unicodedata

def is_english(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

def save_subfolder_names_as_text(root_folder):
    for foldername, subfolders, filenames in os.walk(root_folder):
        for subfolder in subfolders:
            subfolder_path = os.path.join(foldername, subfolder)
            if not is_english(subfolder):

                # https://1023labs.com/posts/python-korean-unicodedata/
                subfolder_join = unicodedata.normalize('NFC', subfolder)

                # 폴더 이름에서 공백을 제거하고, 각 글자를 줄바꿈 문자와 함께 조합
                formatted_text = '\n'.join(list(subfolder_join.replace(' ', '')))
                formatted_text = 'sil\n' + formatted_text + '\nsil'
                # print(formatted_text)

                # 파일에 저장
                filename = f"{subfolder.replace(' ', '')}.align"
                align_path = os.path.join(subfolder_path, '..', '..', 'alignments')
                os.makedirs(align_path, exist_ok=True) 
                filepath = os.path.join(align_path, filename)
                
                with open(filepath, 'w') as file:
                    file.write(formatted_text)
```

```python
# kor
start_unicode = ord('가')  # '가'의 유니코드 값
end_unicode = ord('힣')    # '힣'의 유니코드 값

vocab_korean = [chr(code) for code in range(start_unicode, end_unicode + 1)]
```

```python
# kor
char_to_num = tf.keras.layers.StringLookup(vocabulary=vocab_korean, oov_token="")
num_to_char = tf.keras.layers.StringLookup(
    vocabulary=char_to_num.get_vocabulary(), oov_token="", invert=True
)

# print(
#     f"The vocabulary is: {char_to_num.get_vocabulary()} "
#     f"(size ={char_to_num.vocabulary_size()})"
# )
```

```python
def load_video(path:str) -> List[float]:
    # print(f'load_video() path: {path}')

    cap = cv2.VideoCapture(path)
    frames = []
    fn = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if fn != 75:
        print(f'Wrong video! frame num: {fn}')

    for _ in range(fn):
        ret, frame = cap.read()
        # print(frame)
        # print(frame.shape) # (50, 100, 3)
        if frame is not None:
            # print(frame)
            # print(frame.shape)
            # frame = frame[190:240,120:220,:]
                
            frame = frame[0:50, 0:100, :]
            frame = tf.image.rgb_to_grayscale(frame)
            # print(frame)
            frames.append(frame)
        else:
            print("load_video() path" + path)
    cap.release()

    mean = tf.math.reduce_mean(frames)
    std = tf.math.reduce_std(tf.cast(frames, tf.float32))
    # print(f'mean: {mean} - std: {std}')
    return tf.cast((frames - mean), tf.float32) / std
```

```python
def load_alignments(path:str) -> List[str]:
    print("load_alignments() path: " + path)

    with open(path, 'r') as f:
        lines = f.readlines()
    tokens = []
    for line in lines:
        line = line.split()
        if line[0] != 'sil':  #.align 파일에서 마지막 컬럼 (sil, bin, blue..) 만 사용된다. 데이터를 만들 때 고려하지 않아도 되는 듯.
            tokens = [*tokens,' ',line[0]]
    # print(tokens) # [' ', 'bin', ' ', 'blue', ' ', 'at', ' ', 'l', ' ', 'six', ' ', 'now']
    tokens = tf.strings.unicode_split(tokens, input_encoding='UTF-8')
    # print(tokens)
    tokens = tf.reshape(tokens, (-1))
    # print(tokens)
    tokens = char_to_num(tokens)[1:]
    # print(tokens)
    return tokens # tf.Tensor([ 2  9 14 39  2 12 21  5 39  1 20 39 12 39 19  9 24 39 14 15 23], shape=(21,), dtype=int64)
```

```python
def load_data(path: str):
    
    path = bytes.decode(path.numpy())
    # print("load_data() path: " + path)
    
    sub_path = path.split('/')[-3]
    # print(f'sub_path: {sub_path}')
    
    data_path = path.split('/')[-2]
    # print("load_data() data_path: " + data_path)
    
    file_name = path.split('/')[-1].split('.')[0]
    # print(f'file_name: {file_name}')
        
    # video_path = os.path.join('data', subpath, f'{folder_name}', 'cropped', f'{file_name}') # for cropped image
    # video_path = os.path.join('data', sub_path, data_path, file_name) # for video
    video_path = path
    alignment_path = os.path.join(BASE_PATH, 'data', sub_path, 'alignments', data_path, f'{file_name}.align')
    
    # print(f'video_path: {video_path}')
    # print(f'alignment_path: {alignment_path}')
    
    frames = load_video(video_path)
    alignments = load_alignments(alignment_path)

    return frames, alignments
```

```python
import os
import unicodedata

def is_korean_syllable(char):
    # 유니코드에서 한글 음절 범위 확인
    return '\uac00' <= char <= '\ud7a3'

def get_complete_korean_characters(folder_path):
    complete_korean_characters = set()

    for root, dirs, files in os.walk(folder_path):
        for file_name in files:
            # 파일 이름에서 확장자 제거
            base_name, extension = os.path.splitext(file_name)

            # 파일 이름에서 한글 음절 추출
            korean_syllables = [char for char in base_name if is_korean_syllable(char)]

            # 자음과 모음이 분리되지 않은 글자 추출
            for syllable in korean_syllables:
                char_name = unicodedata.name(syllable)
                if 'JAMO' not in char_name:
                    complete_korean_characters.add(syllable)

    return list(complete_korean_characters)

# 특정 폴더 경로 설정
target_folder = '/Users/giy/Projects/fwo_lipnet_plus/data/kor/alignments/s1/'

# 자음과 모음이 분리되지 않은 한글 글자 리스트 얻기
complete_korean_characters_list = get_complete_korean_characters(target_folder)

# 결과 출력
print(complete_korean_characters_list)
```
